```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from functools import reduce
import pdb
from sklearn.model_selection import train_test_split
#autoML Class
import sys
sys.path.append("Z:\\mamuet\\code\\benchmarking\\ABBautoML")
from ABBtpot.tpot.autoRegressor import ABBautoRegressor
```

```python
datapath = 'Z:\\mamuet\\data\\gef_load_2012'
```

```python
benchmark = pd.read_csv(datapath+'\\Benchmark.csv',dtype=float)
load_history = pd.read_csv(datapath+'\\Load_history.csv.zip',dtype=float,thousands=',')
submission_template = pd.read_csv(datapath+'\\submission_template.csv',dtype=float)
temperature_history = pd.read_csv(datapath+'\\temperature_history.csv.zip',dtype=float)
test = pd.read_csv(datapath+'\\test.csv',dtype=float)
weights = pd.read_csv(datapath+'\\weights.csv',dtype=float)
```

```python
#get the forecasts for every date from all stations
temperature_pivot = temperature_history.pivot(index = ['year','month','day'],columns='station_id')
temperature_pivot.reset_index(inplace=True)
```

```python
temperature_pivot.rename(columns={( 'year', ''):'year',('month', ''):'month',(  'day', ''):'day'},
                        inplace=True)
train_x = pd.merge(load_history[['zone_id', 'year', 'month', 'day']],temperature_pivot,how='left',
                   on=['year','month','day'],
                   suffixes=('','_temp'))
train_y = load_history[[c for c in load_history if c.startswith('h')]]
# cols = train_x.columns
# train_x.loc[:,'ts_id'] = np.arange(len(train_x))
# train_x = train_x[['ts_id']+cols.tolist()]
```

```python
test_x = pd.merge(test,temperature_pivot,how='left',
                 on=['year','month','day'],
                  suffixes=('','_temp')).drop(['id','weight'],axis=1)
```

# train/test split

```python
X_train, X_val, y_train, y_val = train_test_split(train_x, train_y,train_size=0.75, test_size=0.25)

print(X_train.shape,X_val.shape,y_train.shape,y_val.shape)
```

# autoML

```python
#parameters
SEED = 0
dl_params={'max_trials':5,'seed':SEED}
ml_params= dict(generations=20,population_size=50,
                max_eval_time_mins=5,random_state=SEED,
                    periodic_checkpoint_folder='.',verbosity=3,max_time_mins=120,
               multi_output=True)
dl_fit_params={'epochs':100,'validation_split':0.2}
scoring_func = 'mean_squared_error' #from sklearn
scoring_func_params=dict(squared=False) #to get rmse

cat_features = ['zone_id','year','month','day']
num_features = [f for f in train_x if f not in cat_features]

feature_extraction_params = dict(num_features=num_features,cat_features=cat_features,
                     quantileTransformerParams=dict(n_quantiles=100,random_state=SEED, 
                                            output_distribution="normal"),n_components=100)
```

```python
%%time
#define and fit model
ar = ABBautoRegressor(ml_params=ml_params,dl_params=dl_params,run_ml=True,run_dl=True,dl_type='tabular')

model,score = ar.fit(X_train,y_train,X_val,y_val,dl_fit_params=dl_fit_params,
                      scoring_func = scoring_func,scoring_func_params=scoring_func_params,
                    feature_extraction_params=feature_extraction_params,
                    feature_extraction=True,feature_selection=True)
```
